```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data1=pd.read_csv("C:/Users/Gavya/Downloads/Mall_Customers.csv")
data1.head(10)
```

```python
data1 = pd.get_dummies(data1, columns=['Gender'])
```

```python
X=data1.values
print("x:",X)
```

```python
scaler=StandardScaler()
fitted=scaler.fit_transform(X)
```

```python
nclusters=3
kmeans=KMeans(n_clusters=nclusters,random_state=42)
kmeans.fit(fitted)
```

```python
import matplotlib.pyplot as plt
plt.scatter(fitted[:,0],fitted[:,1],c=kmeans.labels_,cmap='viridis')
centers=kmeans.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='black',marker='o',s=200,alpha=0.5)
plt.show()
```
